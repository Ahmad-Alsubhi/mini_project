```python
import matplotlib.pyplot as plt
import pandas as pd

# تحويل عمود 'date' إلى نوع datetime إذا لم يكن كذلك
df = pd.read_csv('sales.csv')
```

```python
df
```

```python
df=df['date','sales']
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# تحويل عمود 'date' إلى نوع datetime
df['date'] = pd.to_datetime(df['date'])

# فرز البيانات حسب التاريخ
df = df.sort_values('date')

# حساب متوسط المبيعات اليومي المتحرك (7 أيام كمثال)
df['sales_moving_avg'] = df['sales'].rolling(window=7).mean()

# رسم السلسلة الزمنية مع المتوسط المتحرك
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['sales'], marker='o', linestyle='-', label='Daily Sales')
plt.plot(df['date'], df['sales_moving_avg'], linestyle='-', linewidth=2, color='red', label='7-Day Moving Average')

# تخصيص الرسم
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Time Series of Sales with Moving Average')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# عرض الرسم
plt.show()
```
